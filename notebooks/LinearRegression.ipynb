# Linear Regression

$$
\hat{y} = w \cdot x + b
$$

## Manual

```python
x = [30, 45, 60, 50, 40, 55, 35, 65, 70, 25]  # Area
y = [150, 200, 250, 220, 180, 240, 160, 270, 290, 130]  # Price
len(x), len(y)
```

Compute means

```python
x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)
x_mean, y_mean
```

Compute slope:

$$
w = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

```python
w = sum((x_i - x_mean) * (y_i - y_mean) for x_i, y_i in zip(x, y)) / sum((x_i - x_mean) ** 2 for x_i in x)
w
```

Compute bias:

$$
b = \bar{y} - w \cdot \bar{x}
$$

```python
b = y_mean - w * x_mean
b
```

Compute predictions

```python
y_pred = [w * x_i + b for x_i in x]
y_pred
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(x, y, c='b', label='Real data')
plt.plot(x, y_pred, c='r', label='Model')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

## Scikit-learn

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array(x).reshape(-1, 1)
y = np.array(y)

model = LinearRegression()
model.fit(X, y)

w = model.coef_[0]
b = model.intercept_

w, b
```

```python
y_pred = model.predict(X)
y_pred
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(x, y, c='b', label='Real data')
plt.plot(x, y_pred, c='r', label='Model')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
